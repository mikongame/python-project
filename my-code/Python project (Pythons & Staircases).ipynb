```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up next to a lake and find yourself in a strange building wich contains grassfields in it and where you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of it and get home, NOW!")
    play_floor(game_state["current_floor"])

def play_floor(floor):
    """
    Play a floor. First check if the floor being played is the target floor.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this floor) or examine an item found here.
    """
    game_state["current_floor"] = floor
    if(game_state["current_floor"] == game_state["target_floor"]):
        print("Congrats! You escaped the floor!")
    else:
        print("You are now in " + floor["name"])
        intended_action = input("What would you like to do? Type 'explore' or 'examine'?").strip()
        if intended_action == "explore":
            explore_floor(floor)
            play_floor(floor)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine?").strip())
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            play_floor(floor)
        linebreak()

def explore_floor(floor):
    """
    Explore a floor. List all items belonging to this floor.
    """
    items = [i["name"] for i in object_relations[floor["name"]]]
    print("You explore the floor. This is " + floor["name"] + ". You find " + ", ".join(items))

def get_next_floor_of_barrier(barrier, current_floor):
    """
    From object_relations, find the two floors connected to the given barrier.
    Return the floor that is not the current_floor.
    """
    connected_floors = object_relations[barrier["name"]]
    for floor in connected_floors:
        if(not current_floor == floor):
            return floor

def examine_item(item_name):
    """
    Examine an item which can be of different types.
    First make sure the intended item belongs to the current floor.
    Then check if the item is a barrier. Tell player if needed item hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    floor. If the item is not a barrier, then check if it contains usable items.
    Collect the item if found and update the game state. At the end,
    play either the current or the next floor depending on the game state
    to keep playing.
    """
    current_floor = game_state["current_floor"]
    next_floor = ""
    output = None
    
    for item in object_relations[current_floor["name"]]:
        if(item["name"] == item_name):
            output = "You examine " + item_name + ". "
            if(item["type"] == "barrier"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You used a item you have. You can now access to the next floor."
                    next_floor = get_next_floor_of_barrier(item, current_floor)
                else:
                    output += "You don't have the item needed to overcome that barrier. Keep looking!"
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            break

    if(output is None):
        print("The item you requested is not found in the current floor.")
    
    if(next_floor and input("Do you want to go to the next floor? Enter 'yes' or 'no'").strip() == 'yes'):
        play_floor(next_floor)
    else:
        play_floor(current_floor)
```
