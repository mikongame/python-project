```python
# define floors and items

#Define floors and staircases
ground_floor = {
    "name": "ground floor",
    "type": "floor",
}

staircase_a = {
    "name": "staircase a",
    "type": "stairs",
}

outside = {
  "name": "outside"
}

#Define non usable objects
lake = {
    "name": "lake",
    "type": "non_usable_object",
}

chest1 = {
    "name": "chest 1",
    "type": "non_usable_object",
}

#Define barriers and enemies
enemy_a = {
    "name": "wild rat",
    "type": "enemy",
}

barrier_a = {
    "name": "magic barrier a",
    "type": "barrier",
}

#Define usable objects
equippable1 = {
    "name": "Short sword",
    "type": "equippable item",
    "target": enemy_a,
}

key_a = {
    "name": "key for magic barrier a",
    "type": "usable item",
    "target": barrier_a,
}

#All things
all_floors = [ground_floor, staircase_a, outside]

all_barriers = [enemy_a, barrier_a]

# define which items/floors are related

object_relations = {
    "ground floor": [lake, chest1, enemy_a],
    "lake": [equippable1],
    "chest 1": [key_a],
    "wild rat": [ground_floor, staircase_a],
    "staircase a": [enemy_a, barrier_a],
    "barrier a": [staircase_a, outside],
    "outside": [barrier_a],
}

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_floor": ground_floor,
    "items_collected": [],
    "target_floor": outside
}
```

```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up on a lake and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!")
    play_floor(game_state["current_floor"])

def play_floor(floor):
    """
    Play a floor. First check if the floor being played is the target floor.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this floor) or examine an item found here.
    """
    game_state["current_floor"] = floor
    if(game_state["current_floor"] == game_state["target_floor"]):
        print("Congrats! You escaped the floor!")
    else:
        print("You are now in " + floor["name"])
        intended_action = input("What would you like to do? Type 'explore' or 'examine'?").strip()
        if intended_action == "explore":
            explore_floor(floor)
            play_floor(floor)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine?").strip())
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            play_floor(floor)
        linebreak()

def explore_floor(floor):
    """
    Explore a floor. List all items belonging to this floor.
    """
    items = [i["name"] for i in object_relations[floor["name"]]]
    print("You explore the floor. This is " + floor["name"] + ". You find " + ", ".join(items))

def get_next_floor_of_barrier(barrier, current_floor):
    """
    From object_relations, find the two floors connected to the given barrier.
    Return the floor that is not the current_floor.
    """
    connected_floors = object_relations[barrier["name"]]
    for floor in connected_floors:
        if(not current_floor == floor):
            return floor

def examine_item(item_name):
    """
    Examine an item which can be a barrier or furniture.
    First make sure the intended item belongs to the current floor.
    Then check if the item is a barrier. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    floor. If the item is not a barrier, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next floor depending on the game state
    to keep playing.
    """
    current_floor = game_state["current_floor"]
    next_floor = ""
    output = None
    
    for item in object_relations[current_floor["name"]]:
        if(item["name"] == item_name):
            output = "You examine " + item_name + ". "
            if(item["type"] == "barrier"):
                have_key = False
                for key in game_state["items_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have."
                    next_floor = get_next_floor_of_barrier(item, current_floor)
                else:
                    output += "It is locked but you don't have the key."
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["items_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            break

    if(output is None):
        print("The item you requested is not found in the current floor.")
    
    if(next_floor and input("Do you want to go to the next floor? Ener 'yes' or 'no'").strip() == 'yes'):
        play_floor(next_floor)
    else:
        play_floor(current_floor)
```
